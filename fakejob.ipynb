```python
stop_words=set(stopwords.words("english"))
```

```python
def clean_text(text):
    if pd.isnull(text):
        return ""
    text=re.sub(r"http\S+|www\S+","",text)
    text=re.sub(r"[^a-z\s]","",text)
    text=text.lower()
    text=" ".join([word for word in text.split() if word not in stop_words])
    return text
```
